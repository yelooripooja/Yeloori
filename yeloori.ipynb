```python
"""
Yeloori Language — Token Definitions
All tokens for the Sanskrit-inspired language.
"""

from enum import Enum, auto


class TokenType(Enum):
    # ── Literals ───────────────────────────────────────────────
    INTEGER     = auto()
    FLOAT       = auto()
    STRING      = auto()
    BOOL_TRUE   = auto()   # satya
    BOOL_FALSE  = auto()   # asatya
    NULL        = auto()   # shoonya

    # ── Identifiers ────────────────────────────────────────────
    IDENTIFIER  = auto()

    # ── Keywords ───────────────────────────────────────────────
    CHARA       = auto()   # var / let
    STHIRA      = auto()   # const
    KARMA       = auto()   # def / function
    PRATYAVARTAYA = auto() # return
    YADI        = auto()   # if
    ANYATHA     = auto()   # else
    ATHAVA      = auto()   # elif
    YAAVAT      = auto()   # while
    CHAKRA      = auto()   # for
    IN          = auto()   # in
    VIRAAM      = auto()   # break
    AGRE        = auto()   # continue
    VARGA       = auto()   # class
    AAHVAAN     = auto()   # import
    SE          = auto()   # from
    LIKHA       = auto()   # print
    PRASHNA     = auto()   # input
    EVAM        = auto()   # pass
    PRAYA       = auto()   # try
    RAKSHA      = auto()   # except
    ANTE        = auto()   # finally
    UTPANNA     = auto()   # raise
    SAHA        = auto()   # with
    YATHA       = auto()   # as
    LAMBDA_KW   = auto()   # sukshma (lambda)
    DELETE_KW   = auto()   # mita (del)
    GLOBAL_KW   = auto()   # vishwa (global)
    ASSERT_KW   = auto()   # nischaya (assert)
    YIELD_KW    = auto()   # pravah (yield)
    ASYNC_KW    = auto()   # asankalan
    AWAIT_KW    = auto()   # pratiksha

    # ── Operators ──────────────────────────────────────────────
    PLUS        = auto()   # +
    MINUS       = auto()   # -
    STAR        = auto()   # *
    SLASH       = auto()   # /
    DOUBLESLASH = auto()   # //
    PERCENT     = auto()   # %
    POWER       = auto()   # **
    EQ          = auto()   # ==
    NEQ         = auto()   # !=
    LT          = auto()   # <
    GT          = auto()   # >
    LTE         = auto()   # <=
    GTE         = auto()   # >=
    ASSIGN      = auto()   # =
    PLUS_EQ     = auto()   # +=
    MINUS_EQ    = auto()   # -=
    STAR_EQ     = auto()   # *=
    SLASH_EQ    = auto()   # /=
    AND         = auto()   # cha (and)
    OR          = auto()   # va  (or)
    NOT         = auto()   # na  (not)

    # ── Delimiters ─────────────────────────────────────────────
    LPAREN      = auto()   # (
    RPAREN      = auto()   # )
    LBRACKET    = auto()   # [
    RBRACKET    = auto()   # ]
    LBRACE      = auto()   # {
    RBRACE      = auto()   # }
    COLON       = auto()   # :
    COMMA       = auto()   # ,
    DOT         = auto()   # .
    ARROW       = auto()   # ->
    SEMICOLON   = auto()   # ;

    # ── Structure ──────────────────────────────────────────────
    NEWLINE     = auto()
    INDENT      = auto()
    DEDENT      = auto()
    EOF         = auto()

    # ── Comments ───────────────────────────────────────────────
    COMMENT     = auto()   # # tippani


class Token:
    """Represents a single token with type, value, and source location."""

    __slots__ = ("type", "value", "line", "col")

    def __init__(self, type_: TokenType, value: object, line: int = 0, col: int = 0):
        self.type  = type_
        self.value = value
        self.line  = line
        self.col   = col

    def __repr__(self):
        return f"Token({self.type.name}, {self.value!r}, {self.line}:{self.col})"

    def __eq__(self, other):
        if isinstance(other, Token):
            return self.type == other.type and self.value == other.value
        return NotImplemented


# ── Keyword map: Sanskrit word → TokenType ────────────────────────────────────
KEYWORDS: dict[str, TokenType] = {
    "chara":          TokenType.CHARA,
    "sthira":         TokenType.STHIRA,
    "karma":          TokenType.KARMA,
    "pratyavartaya":  TokenType.PRATYAVARTAYA,
    "yadi":           TokenType.YADI,
    "anyatha":        TokenType.ANYATHA,
    "athava":         TokenType.ATHAVA,
    "yaavat":         TokenType.YAAVAT,
    "chakra":         TokenType.CHAKRA,
    "in":             TokenType.IN,
    "viraam":         TokenType.VIRAAM,
    "agre":           TokenType.AGRE,
    "varga":          TokenType.VARGA,
    "aahvaan":        TokenType.AAHVAAN,
    "se":             TokenType.SE,
    "likha":          TokenType.LIKHA,
    "prashna":        TokenType.PRASHNA,
    "evam":           TokenType.EVAM,
    "praya":          TokenType.PRAYA,
    "raksha":         TokenType.RAKSHA,
    "ante":           TokenType.ANTE,
    "utpanna":        TokenType.UTPANNA,
    "saha":           TokenType.SAHA,
    "yatha":          TokenType.YATHA,
    "sukshma":        TokenType.LAMBDA_KW,
    "mita":           TokenType.DELETE_KW,
    "vishwa":         TokenType.GLOBAL_KW,
    "nischaya":       TokenType.ASSERT_KW,
    "pravah":         TokenType.YIELD_KW,
    "asankalan":      TokenType.ASYNC_KW,
    "pratiksha":      TokenType.AWAIT_KW,
    "satya":          TokenType.BOOL_TRUE,
    "asatya":         TokenType.BOOL_FALSE,
    "shoonya":        TokenType.NULL,
    "cha":            TokenType.AND,
    "va":             TokenType.OR,
    "na":             TokenType.NOT,
}

# ── Reverse map: TokenType → Sanskrit keyword ─────────────────────────────────
KEYWORD_NAMES: dict[TokenType, str] = {v: k for k, v in KEYWORDS.items()}
```

```python
# Yeloori Language — Lexer
# Converts raw Yeloori source text into a stream of tokens.
# Handles: indentation, strings, numbers, operators, keywords, comments.

import re
from typing import Iterator
# The Token, TokenType, and KEYWORDS are defined in the previous cell and are globally available.
# from .tokens import Token, TokenType, KEYWORDS


class LooriLexError(Exception):
    """Custom exception for lexing errors."""
    def __init__(self, message, line, col, filename):
        super().__init__(f"{filename}:{line}:{col}: Lexing error: {message}")
        self.line = line
        self.col = col
        self.filename = filename


class Lexer:
    """
    Stateful lexer for the Yeloori language.

    Usage:
        lex    = Lexer(source_code)
        tokens = lex.tokenize()
    """

    def __init__(self, source: str, filename: str = "<stdin>"):
        self.source   = source
        self.filename = filename
        self.pos      = 0
        self.line     = 1
        self.col      = 1
        self.tokens: list[Token] = []
        self._indent_stack: list[int] = [0]

    # ── Public API ────────────────────────────────────────────────────────────

    def tokenize(self) -> list[Token]:
        """Tokenize the entire source and return the token list."""
        for tok in self._scan_all():
            self.tokens.append(tok)
        return self.tokens

    # ── Internal scanner ──────────────────────────────────────────────────────

    def _scan_all(self) -> Iterator[Token]:
        lines = self.source.splitlines(keepends=True)

        for lineno, raw_line in enumerate(lines, start=1):
            self.line = lineno
            self.col  = 1
            line      = raw_line.rstrip("\n\r")

            # Skip blank lines and full-line comments
            stripped = line.lstrip()
            if not stripped or stripped.startswith("#"):
                continue

            # ── Indentation ────────────────────────────────────────────────
            indent = len(line) - len(line.lstrip(" "))
            current_indent = self._indent_stack[-1]

            if indent > current_indent:
                self._indent_stack.append(indent)
                yield Token(TokenType.INDENT, indent, lineno, 1)
            elif indent < current_indent:
                while self._indent_stack[-1] > indent:
                    self._indent_stack.pop()
                    yield Token(TokenType.DEDENT, indent, lineno, 1)
                if self._indent_stack[-1] != indent:
                    raise LooriLexError(
                        f"Inconsistent indentation", lineno, 1, self.filename
                    )

            # ── Scan tokens in this line ────────────────────────────────────
            self.col = indent + 1
            pos      = indent
            content  = line

            while pos < len(content):
                ch = content[pos]

                # Whitespace
                if ch in " \t":
                    pos += 1
                    self.col += 1
                    continue

                # Comment
                if ch == "#":
                    break

                # String literals
                if ch in ('"', "'"):
                    tok, advance = self._read_string(content, pos, lineno)
                    yield tok
                    pos      += advance
                    self.col += advance
                    continue

                # Numbers
                if ch.isdigit() or (ch == "-" and pos + 1 < len(content) and content[pos + 1].isdigit()):
                    tok, advance = self._read_number(content, pos, lineno)
                    yield tok
                    pos      += advance
                    self.col += advance
                    continue

                # Identifiers / keywords
                if ch.isalpha() or ch == "_":
                    tok, advance = self._read_identifier(content, pos, lineno)
                    yield tok
                    pos      += advance
                    self.col += advance
                    continue

                # Operators and punctuation
                tok, advance = self._read_operator(content, pos, lineno)
                if tok:
                    yield tok
                    pos      += advance
                    self.col += advance
                    continue

                raise LooriLexError(
                    f"Unknown character: {ch!r}", lineno, self.col, self.filename
                )

            yield Token(TokenType.NEWLINE, "\n", lineno, len(content) + 1)

        # Close remaining indents
        while len(self._indent_stack) > 1:
            self._indent_stack.pop()
            yield Token(TokenType.DEDENT, 0, self.line, 1)

        yield Token(TokenType.EOF, None, self.line, self.col)

    # ── Readers ───────────────────────────────────────────────────────────────

    def _read_string(self, line: str, pos: int, lineno: int):
        quote = line[pos]
        # Triple-quoted strings
        if line[pos:pos+3] in ('"""', "'''"):
            triple = line[pos:pos+3]
            end    = line.find(triple, pos + 3)
            if end == -1:
                raise LooriLexError("Unterminated triple-quoted string", lineno, self.col, self.filename)
            value   = line[pos + 3 : end]
            advance = end + 3 - pos
            return Token(TokenType.STRING, value, lineno, self.col), advance
        # Single-quoted
        end = pos + 1
        result = []
        while end < len(line):
            c = line[end]
            if c == "\\":
                escape = line[end + 1] if end + 1 < len(line) else ""
                ESC = {"n": "\n", "t": "\t", "r": "\r", "\\": "\\", "'": "'", '"': '"'}
                result.append(ESC.get(escape, escape))
                end += 2
            elif c == quote:
                end += 1
                break
            else:
                result.append(c)
                end += 1
        else:
            raise LooriLexError("Unterminated string literal", lineno, self.col, self.filename)
        return Token(TokenType.STRING, "".join(result), lineno, self.col), end - pos

    def _read_number(self, line: str, pos: int, lineno: int):
        match = re.match(r"(0x[0-9A-Fa-f]+|0b[01]+|0o[0-7]+|\d+\.\d*([eE][+-]?\d+)?|\d+([eE][+-]?\d+)?|\d+)", line[pos:])
        if not match:
            raise LooriLexError("Invalid number literal", lineno, self.col, self.filename)
        raw     = match.group()
        advance = len(raw)
        if "." in raw or "e" in raw.lower():
            return Token(TokenType.FLOAT, float(raw), lineno, self.col), advance
        elif raw.startswith("0x"):
            return Token(TokenType.INTEGER, int(raw, 16), lineno, self.col), advance
        elif raw.startswith("0b"):
            return Token(TokenType.INTEGER, int(raw, 2), lineno, self.col), advance
        elif raw.startswith("0o"):
            return Token(TokenType.INTEGER, int(raw, 8), lineno, self.col), advance
        else:
            return Token(TokenType.INTEGER, int(raw), lineno, self.col), advance

    def _read_identifier(self, line: str, pos: int, lineno: int):
        match = re.match(r"[A-Za-z_][A-Za-z0-9_]*", line[pos:])
        word  = match.group()
        ttype = KEYWORDS.get(word, TokenType.IDENTIFIER)
        return Token(ttype, word, lineno, self.col), len(word)

    def _read_operator(self, line: str, pos: int, lineno: int):
        # Two-character operators first
        two = line[pos:pos+2]
        TWO_CHAR = {
            "==": TokenType.EQ,
            "!=": TokenType.NEQ,
            "<=": TokenType.LTE,
            ">=": TokenType.GTE,
            "->": TokenType.ARROW,
            "//": TokenType.DOUBLESLASH,
            "**": TokenType.POWER,
            "+=": TokenType.PLUS_EQ,
            "-=": TokenType.MINUS_EQ,
            "*=": TokenType.STAR_EQ,
            "/=": TokenType.SLASH_EQ,
        }
        if two in TWO_CHAR:
            return Token(TWO_CHAR[two], two, lineno, self.col), 2

        # Single-character operators
        ONE_CHAR = {
            "+": TokenType.PLUS,
            "-": TokenType.MINUS,
            "*": TokenType.STAR,
            "/": TokenType.SLASH,
            "%": TokenType.PERCENT,
            "<": TokenType.LT,
            ">": TokenType.GT,
            "=": TokenType.ASSIGN,
            "(": TokenType.LPAREN,
            ")": TokenType.RPAREN,
            "[": TokenType.LBRACKET,
            "]": TokenType.RBRACKET,
            "{": TokenType.LBRACE,
            "}": TokenType.RBRACE,
            ":": TokenType.COLON,
            ",": TokenType.COMMA,
            ".": TokenType.DOT,
            ";": TokenType.SEMICOLON,
        }
        ch = line[pos]
        if ch in ONE_CHAR:
            return Token(ONE_CHAR[ch], ch, lineno, self.col), 1

        return None, 1
```

```python
# Loori Language — Error System
# All custom exception classes for the Loori runtime.

from __future__ import annotations


class LooriError(Exception):
    """Base class for all Loori errors."""

    def __init__(self, message: str, line: int = 0, col: int = 0, filename: str = "<stdin>"):
        self.message  = message
        self.line     = line
        self.col      = col
        self.filename = filename
        super().__init__(self._format())

    def _format(self) -> str:
        loc = f"{self.filename}:{self.line}:{self.col}" if self.line else self.filename
        return f"\n  ╔══ Loori Dosha (Error) ════════════════════════╗\n  ║  {loc}\n  ║  {self.__class__.__name__}: {self.message}\n  ╚════════════════════════════════════════════════════╝"


class LooriLexError(LooriError):
    """Raised by the Lexer for invalid tokens."""
    pass


class LooriParseError(LooriError):
    """Raised by the Parser for syntax errors."""
    pass


class LooriRuntimeError(LooriError):
    """Raised during interpretation/execution."""
    pass


class LooriTypeError(LooriRuntimeError):
    """Type mismatch errors."""
    pass


class LooriNameError(LooriRuntimeError):
    """Undefined variable or name."""
    pass


class LooriImportError(LooriRuntimeError):
    """Module import failures."""
    pass


class LooriReturnSignal(Exception):
    """Internal signal for 'pratyavartaya' (return) statement."""
    def __init__(self, value):
        self.value = value


class LooriBreakSignal(Exception):
    """Internal signal for 'viraam' (break) statement."""
    pass


class LooriContinueSignal(Exception):
    """Internal signal for 'agre' (continue) statement."""
    pass


class LooriYieldSignal(Exception):
    """Internal signal for 'pravah' (yield) statement."""
    def __init__(self, value):
        self.value = value
```

```python
# Yeloori Language — Abstract Syntax Tree (AST) Nodes
# Every node type in the AST is defined here as a dataclass.

from __future__ import annotations
from dataclasses import dataclass, field
from typing import Any, Optional


# ── Base ──────────────────────────────────────────────────────────────────────

@dataclass(kw_only=True)
class ASTNode:
    """Base class for all AST nodes."""
    line: int = 0
    col:  int = 0


# ── Statements ────────────────────────────────────────────────────────────────

@dataclass
class Program(ASTNode):
    body: list[ASTNode] = field(default_factory=list)


@dataclass
class VarDecl(ASTNode):
    name:    str
    value:   Optional[ASTNode]
    const:   bool = False


@dataclass
class Assign(ASTNode):
    target: ASTNode
    value:  ASTNode
    op:     str = "="  # =, +=, -=, *=, /=


@dataclass
class Block(ASTNode):
    body: list[ASTNode] = field(default_factory=list)


@dataclass
class IfStmt(ASTNode):
    condition: ASTNode
    then_body: Block
    elif_clauses: list[tuple[ASTNode, Block]] = field(default_factory=list)
    else_body: Optional[Block] = None


@dataclass
class WhileStmt(ASTNode):
    condition: ASTNode
    body:      Block


@dataclass
class ForStmt(ASTNode):
    var:      str
    iterable: ASTNode
    body:     Block


@dataclass
class FuncDef(ASTNode):
    name:     str
    params:   list[Param]
    body:     Block
    returns:  Optional[ASTNode] = None
    is_async: bool = False


@dataclass
class Param(ASTNode):
    name:    str
    default: Optional[ASTNode] = None
    type_hint: Optional[str]   = None


@dataclass
class ClassDef(ASTNode):
    name:    str
    base:    Optional[str]
    body:    Block


@dataclass
class ReturnStmt(ASTNode):
    value: Optional[ASTNode]


@dataclass
class BreakStmt(ASTNode):
    pass


@dataclass
class ContinueStmt(ASTNode):
    pass


@dataclass
class PassStmt(ASTNode):
    pass


@dataclass
class ImportStmt(ASTNode):
    module: str
    alias:  Optional[str] = None


@dataclass
class FromImport(ASTNode):
    module:  str
    names:   list[tuple[str, Optional[str]]]  # (name, alias)


@dataclass
class PrintStmt(ASTNode):
    args: list[ASTNode]
    sep:  str = " "
    end:  str = "\n"


@dataclass
class ExprStmt(ASTNode):
    expr: ASTNode


@dataclass
class TryStmt(ASTNode):
    body:       Block
    handlers:   list[ExceptHandler]
    finally_:   Optional[Block] = None
    else_body:  Optional[Block] = None


@dataclass
class ExceptHandler(ASTNode):
    exc_type: Optional[str]
    name:     Optional[str]
    body:     Block


@dataclass
class RaiseStmt(ASTNode):
    exc: Optional[ASTNode]


@dataclass
class WithStmt(ASTNode):
    context: ASTNode
    alias:   Optional[str]
    body:    Block


@dataclass
class GlobalStmt(ASTNode):
    names: list[str]


@dataclass
class AssertStmt(ASTNode):
    test:    ASTNode
    message: Optional[ASTNode]


@dataclass
class DeleteStmt(ASTNode):
    target: ASTNode


@dataclass
class YieldStmt(ASTNode):
    value: Optional[ASTNode]


# ── Expressions ───────────────────────────────────────────────────────────────

@dataclass
class BinOp(ASTNode):
    left:  ASTNode
    op:    str
    right: ASTNode


@dataclass
class UnaryOp(ASTNode):
    op:    str
    right: ASTNode


@dataclass
class BoolOp(ASTNode):
    op:      str          # "cha" (and) | "va" (or)
    values:  list[ASTNode]


@dataclass
class Compare(ASTNode):
    left:  ASTNode
    ops:   list[str]
    right: list[ASTNode]


@dataclass
class Call(ASTNode):
    func:    ASTNode
    args:    list[ASTNode]
    kwargs:  dict[str, ASTNode] = field(default_factory=dict)


@dataclass
class Attribute(ASTNode):
    obj:  ASTNode
    name: str


@dataclass
class Subscript(ASTNode):
    obj:   ASTNode
    index: ASTNode


@dataclass
class Slice(ASTNode):
    start: Optional[ASTNode]
    stop:  Optional[ASTNode]
    step:  Optional[ASTNode]


@dataclass
class Name(ASTNode):
    id: str


@dataclass
class Literal(ASTNode):
    value: Any


@dataclass
class ListExpr(ASTNode):
    elements: list[ASTNode]


@dataclass
class DictExpr(ASTNode):
    keys:   list[ASTNode]
    values: list[ASTNode]


@dataclass
class SetExpr(ASTNode):
    elements: list[ASTNode]


@dataclass
class TupleExpr(ASTNode):
    elements: list[ASTNode]


@dataclass
class Lambda(ASTNode):
    params: list[Param]
    body:   ASTNode


@dataclass
class IfExpr(ASTNode):
    """Ternary: value yadi condition anyatha alt"""
    condition: ASTNode
    then_val:  ASTNode
    else_val:  ASTNode


@dataclass
class ListComp(ASTNode):
    elt:        ASTNode
    var:        str
    iterable:   ASTNode
    condition:  Optional[ASTNode] = None


@dataclass
class Await(ASTNode):
    value: ASTNode


@dataclass
class FString(ASTNode):
    """f-string / formatted string"""
    parts: list[ASTNode | str]
```

```python
# Yeloori Language — Recursive Descent Parser
# Converts a token stream into a full AST.

from __future__ import annotations
from typing import Optional
# Token, TokenType, ASTNode classes, and LooriParseError are defined in previous cells and are globally available.


class Parser:
    """
    Recursive descent parser for Yeloori.

    Usage:
        parser = Parser(tokens)
        ast    = parser.parse()
    """

    def __init__(self, tokens: list[Token], filename: str = "<stdin>"):
        self.tokens   = [t for t in tokens if t.type != TokenType.COMMENT]
        self.pos      = 0
        self.filename = filename

    # ── Token utilities ───────────────────────────────────────────────────────

    def _current(self) -> Token:
        return self.tokens[self.pos] if self.pos < len(self.tokens) else Token(TokenType.EOF, None)

    def _peek(self, offset: int = 1) -> Token:
        idx = self.pos + offset
        return self.tokens[idx] if idx < len(self.tokens) else Token(TokenType.EOF, None)

    def _advance(self) -> Token:
        tok = self._current()
        self.pos += 1
        return tok

    def _check(self, *types: TokenType) -> bool:
        return self._current().type in types

    def _match(self, *types: TokenType) -> bool:
        if self._check(*types):
            self._advance()
            return True
        return False

    def _expect(self, type_: TokenType) -> Token:
        if self._check(type_):
            return self._advance()
        tok = self._current()
        raise LooriParseError(
            f"Expected {type_.name}, got {tok.type.name} ({tok.value!r})",
            tok.line, tok.col, self.filename
        )

    def _skip_newlines(self):
        while self._check(TokenType.NEWLINE):
            self._advance()

    def _at_end(self) -> bool:
        return self._current().type == TokenType.EOF

    # ── Top-level ─────────────────────────────────────────────────────────────

    def parse(self) -> Program:
        self._skip_newlines()
        body = []
        while not self._at_end():
            stmt = self._parse_statement()
            if stmt:
                body.append(stmt)
            self._skip_newlines()
        return Program(body=body)

    # ── Statements ────────────────────────────────────────────────────────────

    def _parse_statement(self) -> Optional[ASTNode]:
        tok = self._current()
        t   = tok.type

        if t == TokenType.NEWLINE:
            self._advance()
            return None
        if t == TokenType.CHARA:
            return self._parse_var_decl(const=False)
        if t == TokenType.STHIRA:
            return self._parse_var_decl(const=True)
        if t == TokenType.KARMA:
            return self._parse_func_def()
        if t == TokenType.ASYNC_KW:
            return self._parse_async_func()
        if t == TokenType.VARGA:
            return self._parse_class_def()
        if t == TokenType.YADI:
            return self._parse_if()
        if t == TokenType.YAAVAT:
            return self._parse_while()
        if t == TokenType.CHAKRA:
            return self._parse_for()
        if t == TokenType.PRATYAVARTAYA:
            return self._parse_return()
        if t == TokenType.VIRAAM:
            self._advance(); self._expect(TokenType.NEWLINE)
            return BreakStmt(line=tok.line, col=tok.col)
        if t == TokenType.AGRE:
            self._advance(); self._expect(TokenType.NEWLINE)
            return ContinueStmt(line=tok.line, col=tok.col)
        if t == TokenType.EVAM:
            self._advance(); self._expect(TokenType.NEWLINE)
            return PassStmt(line=tok.line, col=tok.col)
        if t == TokenType.AAHVAAN:
            return self._parse_import()
        if t == TokenType.SE:
            return self._parse_from_import()
        if t == TokenType.LIKHA:
            return self._parse_print()
        if t == TokenType.PRAYA:
            return self._parse_try()
        if t == TokenType.UTPANNA:
            return self._parse_raise()
        if t == TokenType.SAHA:
            return self._parse_with()
        if t == TokenType.GLOBAL_KW:
            return self._parse_global()
        if t == TokenType.ASSERT_KW:
            return self._parse_assert()
        if t == TokenType.DELETE_KW:
            return self._parse_delete()
        if t == TokenType.YIELD_KW:
            return self._parse_yield()

        return self._parse_expr_stmt()

    def _parse_var_decl(self, const: bool) -> VarDecl:
        tok = self._advance()
        name = self._expect(TokenType.IDENTIFIER).value
        value = None
        if self._match(TokenType.ASSIGN):
            value = self._parse_expression()
        self._expect(TokenType.NEWLINE)
        return VarDecl(name=name, value=value, const=const, line=tok.line, col=tok.col)

    def _parse_func_def(self) -> FuncDef:
        tok = self._advance()
        name = self._expect(TokenType.IDENTIFIER).value
        self._expect(TokenType.LPAREN)
        params = self._parse_params()
        self._expect(TokenType.RPAREN)
        ret = None
        if self._match(TokenType.ARROW):
            ret = self._parse_expression()
        self._expect(TokenType.COLON)
        self._expect(TokenType.NEWLINE)
        body = self._parse_block()
        return FuncDef(name=name, params=params, body=body, returns=ret, line=tok.line, col=tok.col)

    def _parse_async_func(self) -> FuncDef:
        self._advance()
        fd = self._parse_func_def()
        fd.is_async = True
        return fd

    def _parse_params(self) -> list[Param]:
        params = []
        while not self._check(TokenType.RPAREN):
            name = self._expect(TokenType.IDENTIFIER).value
            hint = None
            if self._match(TokenType.COLON):
                hint = self._expect(TokenType.IDENTIFIER).value
            default = None
            if self._match(TokenType.ASSIGN):
                default = self._parse_expression()
            params.append(Param(name=name, default=default, type_hint=hint))
            if not self._match(TokenType.COMMA):
                break
        return params

    def _parse_class_def(self) -> ClassDef:
        tok = self._advance()
        name = self._expect(TokenType.IDENTIFIER).value
        base = None
        if self._match(TokenType.LPAREN):
            base = self._expect(TokenType.IDENTIFIER).value
            self._expect(TokenType.RPAREN)
        self._expect(TokenType.COLON)
        self._expect(TokenType.NEWLINE)
        body = self._parse_block()
        return ClassDef(name=name, base=base, body=body, line=tok.line, col=tok.col)

    def _parse_if(self) -> IfStmt:
        tok = self._advance()
        condition = self._parse_expression()
        self._expect(TokenType.COLON)
        self._expect(TokenType.NEWLINE)
        then_body = self._parse_block()

        elif_clauses = []
        else_body    = None

        while self._check(TokenType.ATHAVA):
            self._advance()
            elif_cond = self._parse_expression()
            self._expect(TokenType.COLON)
            self._expect(TokenType.NEWLINE)
            elif_body = self._parse_block()
            elif_clauses.append((elif_cond, elif_body))

        if self._check(TokenType.ANYATHA):
            self._advance()
            self._expect(TokenType.COLON)
            self._expect(TokenType.NEWLINE)
            else_body = self._parse_block()

        return IfStmt(condition=condition, then_body=then_body,
                      elif_clauses=elif_clauses, else_body=else_body,
                      line=tok.line, col=tok.col)

    def _parse_while(self) -> WhileStmt:
        tok = self._advance()
        condition = self._parse_expression()
        self._expect(TokenType.COLON)
        self._expect(TokenType.NEWLINE)
        body = self._parse_block()
        return WhileStmt(condition=condition, body=body, line=tok.line, col=tok.col)

    def _parse_for(self) -> ForStmt:
        tok = self._advance()
        var = self._expect(TokenType.IDENTIFIER).value
        self._expect(TokenType.IN)
        iterable = self._parse_expression()
        self._expect(TokenType.COLON)
        self._expect(TokenType.NEWLINE)
        body = self._parse_block()
        return ForStmt(var=var, iterable=iterable, body=body, line=tok.line, col=tok.col)

    def _parse_return(self) -> ReturnStmt:
        tok = self._advance()
        value = None
        if not self._check(TokenType.NEWLINE) and not self._at_end():
            value = self._parse_expression()
        if self._check(TokenType.NEWLINE):
            self._advance()
        return ReturnStmt(value=value, line=tok.line, col=tok.col)

    def _parse_import(self) -> ImportStmt:
        tok = self._advance()
        module = self._expect(TokenType.IDENTIFIER).value
        while self._match(TokenType.DOT):
            module += "." + self._expect(TokenType.IDENTIFIER).value
        alias = None
        if self._match(TokenType.YATHA):
            alias = self._expect(TokenType.IDENTIFIER).value
        if self._check(TokenType.NEWLINE):
            self._advance()
        return ImportStmt(module=module, alias=alias, line=tok.line, col=tok.col)

    def _parse_from_import(self) -> FromImport:
        tok = self._advance()
        module = self._expect(TokenType.IDENTIFIER).value
        self._expect(TokenType.AAHVAAN)
        names = []
        while True:
            name = self._expect(TokenType.IDENTIFIER).value
            alias = None
            if self._match(TokenType.YATHA):
                alias = self._expect(TokenType.IDENTIFIER).value
            names.append((name, alias))
            if not self._match(TokenType.COMMA):
                break
        if self._check(TokenType.NEWLINE):
            self._advance()
        return FromImport(module=module, names=names, line=tok.line, col=tok.col)

    def _parse_print(self) -> PrintStmt:
        tok = self._advance()
        self._expect(TokenType.LPAREN)
        args = []
        sep, end = " ", "\n"
        while not self._check(TokenType.RPAREN):
            if self._check(TokenType.IDENTIFIER) and self._peek().type == TokenType.ASSIGN:
                kw = self._advance().value
                self._advance()
                val = self._parse_expression()
                if kw == "sep":
                    sep = val.value if isinstance(val, Literal) else " "
                elif kw == "end":
                    end = val.value if isinstance(val, Literal) else "\n"
            else:
                args.append(self._parse_expression())
            if not self._match(TokenType.COMMA):
                break
        self._expect(TokenType.RPAREN)
        if self._check(TokenType.NEWLINE):
            self._advance()
        return PrintStmt(args=args, sep=sep, end=end, line=tok.line, col=tok.col)

    def _parse_try(self) -> TryStmt:
        tok = self._advance()
        self._expect(TokenType.COLON)
        self._expect(TokenType.NEWLINE)
        body     = self._parse_block()
        handlers = []
        else_body   = None
        finally_ = None

        while self._check(TokenType.RAKSHA):
            self._advance()
            exc_type = None
            name     = None
            if not self._check(TokenType.COLON):
                exc_type = self._expect(TokenType.IDENTIFIER).value
                if self._match(TokenType.YATHA):
                    name = self._expect(TokenType.IDENTIFIER).value
            self._expect(TokenType.COLON)
            self._expect(TokenType.NEWLINE)
            hbody = self._parse_block()
            handlers.append(ExceptHandler(exc_type=exc_type, name=name, body=hbody))

        if self._check(TokenType.ANTE):
            self._advance()
            self._expect(TokenType.COLON)
            self._expect(TokenType.NEWLINE)
            finally_ = self._parse_block()

        return TryStmt(body=body, handlers=handlers, finally_=finally_,
                       else_body=else_body, line=tok.line, col=tok.col)

    def _parse_raise(self) -> RaiseStmt:
        tok = self._advance()
        exc = None
        if not self._check(TokenType.NEWLINE):
            exc = self._parse_expression()
        if self._check(TokenType.NEWLINE):
            self._advance()
        return RaiseStmt(exc=exc, line=tok.line, col=tok.col)

    def _parse_with(self) -> WithStmt:
        tok = self._advance()
        context = self._parse_expression()
        alias   = None
        if self._match(TokenType.YATHA):
            alias = self._expect(TokenType.IDENTIFIER).value
        self._expect(TokenType.COLON)
        self._expect(TokenType.NEWLINE)
        body = self._parse_block()
        return WithStmt(context=context, alias=alias, body=body, line=tok.line, col=tok.col)

    def _parse_global(self) -> GlobalStmt:
        tok = self._advance()
        names = [self._expect(TokenType.IDENTIFIER).value]
        while self._match(TokenType.COMMA):
            names.append(self._expect(TokenType.IDENTIFIER).value)
        if self._check(TokenType.NEWLINE):
            self._advance()
        return GlobalStmt(names=names, line=tok.line, col=tok.col)

    def _parse_assert(self) -> AssertStmt:
        tok = self._advance()
        test = self._parse_expression()
        msg  = None
        if self._match(TokenType.COMMA):
            msg = self._parse_expression()
        if self._check(TokenType.NEWLINE):
            self._advance()
        return AssertStmt(test=test, message=msg, line=tok.line, col=tok.col)

    def _parse_delete(self) -> DeleteStmt:
        tok = self._advance()
        target = self._parse_expression()
        if self._check(TokenType.NEWLINE):
            self._advance()
        return DeleteStmt(target=target, line=tok.line, col=tok.col)

    def _parse_yield(self) -> YieldStmt:
        tok = self._advance()
        value = None
        if not self._check(TokenType.NEWLINE):
            value = self._parse_expression()
        if self._check(TokenType.NEWLINE):
            self._advance()
        return YieldStmt(value=value, line=tok.line, col=tok.col)

    def _parse_expr_stmt(self) -> ASTNode:
        tok  = self._current()
        expr = self._parse_expression()

        # Augmented assignment
        AUG = {
            TokenType.PLUS_EQ:  "+=",
            TokenType.MINUS_EQ: "-=",
            TokenType.STAR_EQ:  "*=",
            TokenType.SLASH_EQ: "/=",
        }
        if self._current().type in AUG:
            op = AUG[self._advance().type]
            val = self._parse_expression()
            if self._check(TokenType.NEWLINE):
                self._advance()
            return Assign(target=expr, value=val, op=op, line=tok.line, col=tok.col)

        # Plain assignment
        if self._match(TokenType.ASSIGN):
            value = self._parse_expression()
            if self._check(TokenType.NEWLINE):
                self._advance()
            return Assign(target=expr, value=value, line=tok.line, col=tok.col)

        if self._check(TokenType.NEWLINE):
            self._advance()
        return ExprStmt(expr=expr, line=tok.line, col=tok.col)

    # ── Block ─────────────────────────────────────────────────────────────────

    def _parse_block(self) -> Block:
        tok = self._current()
        self._expect(TokenType.INDENT)
        stmts = []
        while not self._check(TokenType.DEDENT) and not self._at_end():
            stmt = self._parse_statement()
            if stmt:
                stmts.append(stmt)
        self._expect(TokenType.DEDENT)
        return Block(body=stmts, line=tok.line, col=tok.col)

    # ── Expressions (precedence climbing) ────────────────────────────────────

    def _parse_expression(self) -> ASTNode:
        return self._parse_or()

    def _parse_or(self) -> ASTNode:
        left = self._parse_and()
        while self._check(TokenType.OR):
            self._advance()
            right = self._parse_and()
            left  = BoolOp(op="va", values=[left, right])
        return left

    def _parse_and(self) -> ASTNode:
        left = self._parse_not()
        while self._check(TokenType.AND):
            self._advance()
            right = self._parse_not()
            left  = BoolOp(op="cha", values=[left, right])
        return left

    def _parse_not(self) -> ASTNode:
        if self._check(TokenType.NOT):
            tok = self._advance()
            return UnaryOp(op="na", right=self._parse_not(), line=tok.line, col=tok.col)
        return self._parse_comparison()

    def _parse_comparison(self) -> ASTNode:
        left = self._parse_additive()
        OPS  = {
            TokenType.EQ:  "==", TokenType.NEQ: "!=",
            TokenType.LT:  "<",  TokenType.GT:  ">",
            TokenType.LTE: "<=", TokenType.GTE: ">=",
            TokenType.IN:  "in",
        }
        ops, comparators = [], []
        while self._current().type in OPS:
            ops.append(OPS[self._advance().type])
            comparators.append(self._parse_additive())
        if ops:
            return Compare(left=left, ops=ops, right=comparators)
        return left

    def _parse_additive(self) -> ASTNode:
        left = self._parse_multiplicative()
        while self._check(TokenType.PLUS, TokenType.MINUS):
            op    = self._advance().value
            right = self._parse_multiplicative()
            left  = BinOp(left=left, op=op, right=right)
        return left

    def _parse_multiplicative(self) -> ASTNode:
        left = self._parse_power()
        while self._check(TokenType.STAR, TokenType.SLASH, TokenType.DOUBLESLASH, TokenType.PERCENT):
            op    = self._advance().value
            right = self._parse_power()
            left  = BinOp(left=left, op=op, right=right)
        return left

    def _parse_power(self) -> ASTNode:
        base = self._parse_unary()
        if self._check(TokenType.POWER):
            self._advance()
            exp = self._parse_unary()
            return BinOp(left=base, op="**", right=exp)
        return base

    def _parse_unary(self) -> ASTNode:
        if self._check(TokenType.MINUS):
            tok = self._advance()
            return UnaryOp(op="-", right=self._parse_unary(), line=tok.line, col=tok.col)
        if self._check(TokenType.PLUS):
            self._advance()
            return self._parse_unary()
        return self._parse_postfix()

    def _parse_postfix(self) -> ASTNode:
        expr = self._parse_primary()
        while True:
            if self._check(TokenType.DOT):
                self._advance()
                attr = self._expect(TokenType.IDENTIFIER).value
                expr = Attribute(obj=expr, name=attr, line=expr.line, col=expr.col)
            elif self._check(TokenType.LPAREN):
                expr = self._parse_call(expr)
            elif self._check(TokenType.LBRACKET):
                self._advance()
                idx  = self._parse_expression()
                self._expect(TokenType.RBRACKET)
                expr = Subscript(obj=expr, index=idx, line=expr.line, col=expr.col)
            else:
                break
        return expr

    def _parse_call(self, func: ASTNode) -> Call:
        self._expect(TokenType.LPAREN)
        args, kwargs = [], {}
        while not self._check(TokenType.RPAREN):
            if self._check(TokenType.IDENTIFIER) and self._peek().type == TokenType.ASSIGN:
                key = self._advance().value
                self._advance()
                kwargs[key] = self._parse_expression()
            else:
                args.append(self._parse_expression())
            if not self._match(TokenType.COMMA):
                break
        self._expect(TokenType.RPAREN)
        return Call(func=func, args=args, kwargs=kwargs, line=func.line, col=func.col)

    def _parse_primary(self) -> ASTNode:
        tok = self._current()

        if tok.type == TokenType.INTEGER:
            self._advance()
            return Literal(value=tok.value, line=tok.line, col=tok.col)
        if tok.type == TokenType.FLOAT:
            self._advance()
            return Literal(value=tok.value, line=tok.line, col=tok.col)
        if tok.type == TokenType.STRING:
            self._advance()
            return Literal(value=tok.value, line=tok.line, col=tok.col)
        if tok.type == TokenType.BOOL_TRUE:
            self._advance()
            return Literal(value=True, line=tok.line, col=tok.col)
        if tok.type == TokenType.BOOL_FALSE:
            self._advance()
            return Literal(value=False, line=tok.line, col=tok.col)
        if tok.type == TokenType.NULL:
            self._advance()
            return Literal(value=None, line=tok.line, col=tok.col)
        if tok.type == TokenType.IDENTIFIER:
            self._advance()
            return Name(id=tok.value, line=tok.line, col=tok.col)
        if tok.type == TokenType.LPAREN:
            self._advance()
            expr = self._parse_expression()
            if self._match(TokenType.COMMA):
                elems = [expr]
                while not self._check(TokenType.RPAREN):
                    elems.append(self._parse_expression())
                    if not self._match(TokenType.COMMA):
                        break
                self._expect(TokenType.RPAREN)
                return TupleExpr(elements=elems, line=tok.line, col=tok.col)
            self._expect(TokenType.RPAREN)
            return expr
        if tok.type == TokenType.LBRACKET:
            return self._parse_list()
        if tok.type == TokenType.LBRACE:
            return self._parse_dict_or_set()
        if tok.type == TokenType.LAMBDA_KW:
            return self._parse_lambda()
        if tok.type == TokenType.LIKHA:
            # likha used as expression (rare)
            self._advance()
            return Name(id="likha", line=tok.line, col=tok.col)

        raise LooriParseError(
            f"Unexpected token: {tok.type.name} ({tok.value!r})",
            tok.line, tok.col, self.filename
        )

    def _parse_list(self) -> ListExpr:
        tok = self._advance()
        elems = []
        while not self._check(TokenType.RBRACKET):
            elems.append(self._parse_expression())
            if not self._match(TokenType.COMMA):
                break
        self._expect(TokenType.RBRACKET)
        return ListExpr(elements=elems, line=tok.line, col=tok.col)

    def _parse_dict_or_set(self) -> ASTNode:
        tok = self._advance()
        if self._check(TokenType.RBRACE):
            self._advance()
            return DictExpr(keys=[], values=[], line=tok.line, col=tok.col)

        first = self._parse_expression()
        if self._match(TokenType.COLON):
            # Dict
            keys, values = [first], [self._parse_expression()]
            while self._match(TokenType.COMMA):
                if self._check(TokenType.RBRACE):
                    break
                keys.append(self._parse_expression())
                self._expect(TokenType.COLON)
                values.append(self._parse_expression())
            self._expect(TokenType.RBRACE)
            return DictExpr(keys=keys, values=values, line=tok.line, col=tok.col)
        else:
            # Set
            elems = [first]
            while self._match(TokenType.COMMA):
                if self._check(TokenType.RBRACE):
                    break
                elems.append(self._parse_expression())
            self._expect(TokenType.RBRACE)
            return SetExpr(elements=elems, line=tok.line, col=tok.col)

    def _parse_lambda(self) -> Lambda:
        tok = self._advance()
        params = []
        if not self._check(TokenType.COLON):
            params = self._parse_params()
        self._expect(TokenType.COLON)
        body = self._parse_expression()
        return Lambda(params=params, body=body, line=tok.line, col=tok.col)
```

```python
"""
Yeloori Language — Tree-Walking Interpreter
Executes the AST directly. Handles all language features.
"""

from __future__ import annotations
import math, time, os, sys, json, random, asyncio
from typing import Any, Optional
# Token, TokenType, and KEYWORDS are defined in the previous cell and are globally available.
# ASTNode classes are defined in a previous cell and are globally available.
# Loori error classes are defined in a previous cell and are globally available.


# ── Environment (Scope) ───────────────────────────────────────────────────────

class Environment:
    """Lexical scope — a linked chain of variable stores."""

    def __init__(self, parent: Optional[Environment] = None, name: str = "<scope>"):
        self._vars:   dict[str, Any] = {}
        self._consts: set[str]       = set()
        self.parent = parent
        self.name   = name

    def get(self, name: str, line: int = 0) -> Any:
        if name in self._vars:
            return self._vars[name]
        if self.parent:
            return self.parent.get(name, line)
        raise LooriNameError(f"'{name}' is not defined", line)

    def set(self, name: str, value: Any, const: bool = False):
        if name in self._consts:
            raise LooriRuntimeError(f"Cannot reassign constant '{name}'")
        self._vars[name] = value
        if const:
            self._consts.add(name)

    def assign(self, name: str, value: Any, line: int = 0):
        """Assign to an existing binding (walk up scope chain)."""
        if name in self._vars:
            if name in self._consts:
                raise LooriRuntimeError(f"Cannot reassign constant '{name}'")
            self._vars[name] = value
            return
        if self.parent:
            self.parent.assign(name, value, line)
            return
        # New binding in current scope
        self._vars[name] = value

    def make_child(self, name: str = "<block>") -> Environment:
        return Environment(parent=self, name=name)


# ── Yeloori Function Object ───────────────────────────────────────────────────

class YelooriFunction:
    def __init__(self, node: FuncDef, closure: Environment, interp: "Interpreter"):
        self.node      = node
        self.closure   = closure
        self.interp    = interp
        self.__name__  = node.name

    def call(self, args: list, kwargs: dict) -> Any:
        env = self.closure.make_child(f"karma:{self.node.name}")
        params = self.node.params
        # Positional args
        for i, param in enumerate(params):
            if i < len(args):
                env.set(param.name, args[i])
            elif param.default is not None:
                env.set(param.name, self.interp.eval_node(param.default, self.closure))
            elif param.name in kwargs:
                env.set(param.name, kwargs[param.name])
            else:
                raise LooriRuntimeError(
                    f"Missing argument '{param.name}' for karma '{self.node.name}'"
                )
        # Keyword args not already set
        for k, v in kwargs.items():
            if k not in [p.name for p in params]:
                raise LooriRuntimeError(f"Unexpected keyword argument '{k}'")
            if not env._vars.get(k):
                env.set(k, v)
        try:
            self.interp.exec_block(self.node.body, env)
        except LooriReturnSignal as ret:
            return ret.value
        return None

    def __repr__(self):
        return f"<karma {self.node.name}>"


# ── Yeloori Class Object ──────────────────────────────────────────────────────

class YelooriClass:
    def __init__(self, node: ClassDef, env: Environment, interp: "Interpreter"):
        self.node   = node
        self.interp = interp
        self.methods: dict[str, YelooriFunction] = {}
        # Execute class body to collect methods
        class_env = env.make_child(f"varga:{node.name}")
        interp.exec_block(node.body, class_env)
        for name, val in class_env._vars.items():
            if isinstance(val, YelooriFunction):
                self.methods[name] = val
        self.__name__ = node.name

    def instantiate(self, args, kwargs) -> "YelooriInstance":
        inst = YelooriInstance(self)
        if "arambha" in self.methods:     # arambha = __init__
            fn = self.methods["arambha"]
            fn.call([inst] + args, kwargs)
        return inst

    def __repr__(self):
        return f"<varga {self.node.name}>"


class YelooriInstance:
    def __init__(self, klass: YelooriClass):
        self.klass  = klass
        self.attrs: dict[str, Any] = {}

    def get_attr(self, name: str) -> Any:
        if name in self.attrs:
            return self.attrs[name]
        if name in self.klass.methods:
            # Bind self
            method = self.klass.methods[name]
            return BoundMethod(method, self)
        raise LooriRuntimeError(f"'{self.klass.node.name}' has no attribute '{name}'")

    def set_attr(self, name: str, value: Any):
        self.attrs[name] = value

    def __repr__(self):
        return f"<{self.klass.node.name} samskara>"


class BoundMethod:
    def __init__(self, fn: YelooriFunction, instance: YelooriInstance):
        self.fn       = fn
        self.instance = instance

    def call(self, args, kwargs):
        return self.fn.call([self.instance] + args, kwargs)

    def __repr__(self):
        return f"<bound karma {self.fn.node.name}>"


# ── Built-in Functions ────────────────────────────────────────────────────────

def _builtin_range(*args):
    if len(args) == 1:
        return range(int(args[0]))
    elif len(args) == 2:
        return range(int(args[0]), int(args[1]))
    elif len(args) == 3:
        return range(int(args[0]), int(args[1]), int(args[2]))
    raise LooriRuntimeError("range() takes 1–3 arguments")

BUILTINS: dict[str, Any] = {
    # Core
    "range":  _builtin_range,
    "len":    len,
    "type":   lambda x: type(x).__name__,
    "str":    str,
    "int":    int,
    "float":  float,
    "bool":   bool,
    "list":   list,
    "dict":   dict,
    "set":    set,
    "tuple":  tuple,
    "abs":    abs,
    "round":  round,
    "min":    min,
    "max":    max,
    "sum":    sum,
    "sorted": sorted,
    "reversed": lambda x: list(reversed(x)),
    "enumerate": enumerate,
    "zip":    zip,
    "map":    map,
    "filter": filter,
    "any":    any,
    "all":    all,
    "hasattr": hasattr,
    "getattr": getattr,
    "setattr": setattr,
    "callable": callable,
    "isinstance": isinstance,
    "id":     id,
    "hex":    hex,
    "bin":    bin,
    "oct":    oct,
    "ord":    ord,
    "chr":    chr,
    "repr":   repr,
    "hash":   hash,
    "open":   open,
    "input":  input,
    "exit":   sys.exit,
    # Math
    "ganita": math,
    # JSON
    "json":   json,
    # OS
    "prakriti": os,     # prakriti = nature = OS
    # Random
    "yadriccha": random,  # yadriccha = randomness
    # Time
    "kaala": time,        # kaala = time
    # Yeloori-specific builtins
    "satya":   True,
    "asatya":  False,
    "shoonya": None,
}


# ── Main Interpreter ──────────────────────────────────────────────────────────

class Interpreter:
    """
    Tree-walking interpreter for Yeloori.

    Usage:
        interp = Interpreter()
        interp.execute(ast)
    """

    def __init__(self):
        self.global_env = Environment(name="<global>")
        self._load_builtins()

    def _load_builtins(self):
        for name, val in BUILTINS.items():
            self.global_env.set(name, val)

    def execute(self, program: Program) -> None:
        for node in program.body:
            self.exec_node(node, self.global_env)

    # ── Statement dispatch ────────────────────────────────────────────────────

    def exec_node(self, node: ASTNode, env: Environment) -> Any:
        method = f"exec_{type(node).__name__}"
        handler = getattr(self, method, None)
        if handler:
            return handler(node, env)
        # Fall through to eval for expressions
        return self.eval_node(node, env)

    def exec_block(self, block: Block, env: Environment) -> None:
        for stmt in block.body:
            self.exec_node(stmt, env)

    def exec_Program(self, node: Program, env: Environment):
        for stmt in node.body:
            self.exec_node(stmt, env)

    def exec_VarDecl(self, node: VarDecl, env: Environment):
        value = self.eval_node(node.value, env) if node.value is not None else None
        env.set(node.name, value, const=node.const)

    def exec_Assign(self, node: Assign, env: Environment):
        value = self.eval_node(node.value, env)
        target = node.target

        if node.op != "=":
            current = self.eval_node(target, env)
            OPS = {"+=": lambda a,b: a+b, "-=": lambda a,b: a-b,
                   "*=": lambda a,b: a*b, "/=": lambda a,b: a/b}
            value = OPS[node.op](current, value)

        if isinstance(target, Name):
            env.assign(target.id, value, target.line)
        elif isinstance(target, Attribute):
            obj = self.eval_node(target.obj, env)
            if isinstance(obj, YelooriInstance):
                obj.set_attr(target.name, value)
            else:
                setattr(obj, target.name, value)
        elif isinstance(target, Subscript):
            obj = self.eval_node(target.obj, env)
            idx = self.eval_node(target.index, env)
            obj[idx] = value
        else:
            raise LooriRuntimeError(f"Invalid assignment target", node.line)

    def exec_ExprStmt(self, node: ExprStmt, env: Environment):
        self.eval_node(node.expr, env)

    def exec_PrintStmt(self, node: PrintStmt, env: Environment):
        values = [self._to_str(self.eval_node(a, env)) for a in node.args]
        print(node.sep.join(values), end=node.end)

    def exec_IfStmt(self, node: IfStmt, env: Environment):
        if self._is_truthy(self.eval_node(node.condition, env)):
            self.exec_block(node.then_body, env.make_child())
            return
        for cond, body in node.elif_clauses:
            if self._is_truthy(self.eval_node(cond, env)):
                self.exec_block(body, env.make_child())
                return
        if node.else_body:
            self.exec_block(node.else_body, env.make_child())

    def exec_WhileStmt(self, node: WhileStmt, env: Environment):
        while self._is_truthy(self.eval_node(node.condition, env)):
            try:
                self.exec_block(node.body, env.make_child())
            except LooriBreakSignal:
                break
            except LooriContinueSignal:
                continue

    def exec_ForStmt(self, node: ForStmt, env: Environment):
        iterable = self.eval_node(node.iterable, env)
        try:
            for item in iterable:
                loop_env = env.make_child()
                loop_env.set(node.var, item)
                try:
                    self.exec_block(node.body, loop_env)
                except LooriContinueSignal:
                    continue
        except LooriBreakSignal:
            pass

    def exec_FuncDef(self, node: FuncDef, env: Environment):
        fn = YelooriFunction(node, env, self)
        env.set(node.name, fn)

    def exec_ClassDef(self, node: ClassDef, env: Environment):
        klass = YelooriClass(node, env, self)
        env.set(node.name, klass)

    def exec_ReturnStmt(self, node: ReturnStmt, env: Environment):
        value = self.eval_node(node.value, env) if node.value else None
        raise LooriReturnSignal(value)

    def exec_BreakStmt(self, node: BreakStmt, env: Environment):
        raise LooriBreakSignal()

    def exec_ContinueStmt(self, node: ContinueStmt, env: Environment):
        raise LooriContinueSignal()

    def exec_PassStmt(self, node: PassStmt, env: Environment):
        pass

    def exec_ImportStmt(self, node: ImportStmt, env: Environment):
        try:
            mod = __import__(node.module, fromlist=[""])
            alias = node.alias or node.module.split(".")[0]
            env.set(alias, mod)
        except ImportError as e:
            raise LooriImportError(f"Cannot import '{node.module}': {e}", node.line)

    def exec_FromImport(self, node: FromImport, env: Environment):
        try:
            mod = __import__(node.module, fromlist=[n for n, _ in node.names])
            for name, alias in node.names:
                val = getattr(mod, name)
                env.set(alias or name, val)
        except (ImportError, AttributeError) as e:
            raise LooriImportError(f"Import error: {e}", node.line)

    def exec_TryStmt(self, node: TryStmt, env: Environment):
        try:
            self.exec_block(node.body, env.make_child())
        except LooriRuntimeError as e:
            for handler in node.handlers:
                handler_env = env.make_child()
                if handler.name:
                    handler_env.set(handler.name, e)
                self.exec_block(handler.body, handler_env)
                break
        except Exception as e:
            for handler in node.handlers:
                handler_env = env.make_child()
                if handler.name:
                    handler_env.set(handler.name, e)
                self.exec_block(handler.body, handler_env)
                break
        finally:
            if node.finally_:
                self.exec_block(node.finally_, env.make_child())

    def exec_RaiseStmt(self, node: RaiseStmt, env: Environment):
        if node.exc:
            exc = self.eval_node(node.exc, env)
            if isinstance(exc, str):
                raise LooriRuntimeError(exc, node.line)
            raise exc
        raise LooriRuntimeError("Raised an error", node.line)

    def exec_WithStmt(self, node: WithStmt, env: Environment):
        ctx = self.eval_node(node.context, env)
        with_env = env.make_child()
        if node.alias:
            with_env.set(node.alias, ctx)
        with ctx:
            self.exec_block(node.body, with_env)

    def exec_GlobalStmt(self, node: GlobalStmt, env: Environment):
        for name in node.names:
            val = self.global_env.get(name)
            env.set(name, val)

    def exec_AssertStmt(self, node: AssertStmt, env: Environment):
        result = self.eval_node(node.test, env)
        if not self._is_truthy(result):
            msg = self.eval_node(node.message, env) if node.message else "Assertion failed"
            raise LooriRuntimeError(str(msg), node.line)

    def exec_DeleteStmt(self, node: DeleteStmt, env: Environment):
        if isinstance(node.target, Name):
            del env._vars[node.target.id]
        elif isinstance(node.target, Subscript):
            obj = self.eval_node(node.target.obj, env)
            idx = self.eval_node(node.target.index, env)
            del obj[idx]

    def exec_YieldStmt(self, node: YieldStmt, env: Environment):
        value = self.eval_node(node.value, env) if node.value else None
        raise LooriYieldSignal(value)

    # ── Expression evaluation ─────────────────────────────────────────────────

    def eval_node(self, node: ASTNode, env: Environment) -> Any:
        method = f"eval_{type(node).__name__}"
        handler = getattr(self, method, None)
        if handler:
            return handler(node, env)
        raise LooriRuntimeError(f"Cannot evaluate node: {type(node).__name__}", getattr(node, 'line', 0))

    def eval_Literal(self, node: Literal, env: Environment) -> Any:
        return node.value

    def eval_Name(self, node: Name, env: Environment) -> Any:
        return env.get(node.id, node.line)

    def eval_BinOp(self, node: BinOp, env: Environment) -> Any:
        L = self.eval_node(node.left, env)
        R = self.eval_node(node.right, env)
        op = node.op
        try:
            if op == "+":  return L + R
            if op == "-":  return L - R
            if op == "*":  return L * R
            if op == "/":  return L / R
            if op == "//": return L // R
            if op == "%":  return L % R
            if op == "**": return L ** R
        except TypeError as e:
            raise LooriTypeError(str(e), node.line)
        except ZeroDivisionError:
            raise LooriRuntimeError("Division by zero (shoonya vibhajan)", node.line)

    def eval_UnaryOp(self, node: UnaryOp, env: Environment) -> Any:
        val = self.eval_node(node.right, env)
        if node.op == "-": return -val
        if node.op == "na": return not self._is_truthy(val)
        return val

    def eval_BoolOp(self, node: BoolOp, env: Environment) -> Any:
        if node.op == "cha":  # and
            result = True
            for v in node.values:
                result = self.eval_node(v, env)
                if not self._is_truthy(result):
                    return result
            return result
        if node.op == "va":  # or
            for v in node.values:
                result = self.eval_node(v, env)
                if self._is_truthy(result):
                    return result
            return result

    def eval_Compare(self, node: Compare, env: Environment) -> Any:
        left = self.eval_node(node.left, env)
        for op, right_node in zip(node.ops, node.right):
            right = self.eval_node(right_node, env)
            result = {
                "==": lambda a,b: a == b,
                "!=": lambda a,b: a != b,
                "<":  lambda a,b: a < b,
                ">":  lambda a,b: a > b,
                "<=": lambda a,b: a <= b,
                ">=": lambda a,b: a >= b,
                "in": lambda a,b: a in b,
            }[op](left, right)
            if not result:
                return False
            left = right
        return True

    def eval_Call(self, node: Call, env: Environment) -> Any:
        func   = self.eval_node(node.func, env)
        args   = [self.eval_node(a, env) for a in node.args]
        kwargs = {k: self.eval_node(v, env) for k, v in node.kwargs.items()}

        if isinstance(func, (YelooriFunction,)):
            return func.call(args, kwargs)
        if isinstance(func, BoundMethod):
            return func.call(args, kwargs)
        if isinstance(func, YelooriClass):
            return func.instantiate(args, kwargs)
        if callable(func):
            try:
                return func(*args, **kwargs)
            except Exception as e:
                raise LooriRuntimeError(str(e), node.line)

        raise LooriTypeError(f"'{func}' is not callable", node.line)

    def eval_Attribute(self, node: Attribute, env: Environment) -> Any:
        obj = self.eval_node(node.obj, env)
        if isinstance(obj, YelooriInstance):
            return obj.get_attr(node.name)
        try:
            return getattr(obj, node.name)
        except AttributeError:
            raise LooriRuntimeError(f"No attribute '{node.name}' on {obj!r}", node.line)

    def eval_Subscript(self, node: Subscript, env: Environment) -> Any:
        obj = self.eval_node(node.obj, env)
        idx = self.eval_node(node.index, env)
        try:
            return obj[idx]
        except (KeyError, IndexError, TypeError) as e:
            raise LooriRuntimeError(str(e), node.line)

    def eval_ListExpr(self, node: ListExpr, env: Environment) -> list:
        return [self.eval_node(e, env) for e in node.elements]

    def eval_DictExpr(self, node: DictExpr, env: Environment) -> dict:
        return {self.eval_node(k, env): self.eval_node(v, env)
                for k, v in zip(node.keys, node.values)}

    def eval_SetExpr(self, node: SetExpr, env: Environment) -> set:
        return {self.eval_node(e, env) for e in node.elements}

    def eval_TupleExpr(self, node: TupleExpr, env: Environment) -> tuple:
        return tuple(self.eval_node(e, env) for e in node.elements)

    def eval_Lambda(self, node: Lambda, env: Environment) -> YelooriFunction:
        # Wrap lambda as anonymous function
        func_node = FuncDef(name="<sukshma>", params=node.params,
                            body=Block(body=[ReturnStmt(value=node.body)]))
        return YelooriFunction(func_node, env, self)

    def eval_IfExpr(self, node: IfExpr, env: Environment) -> Any:
        if self._is_truthy(self.eval_node(node.condition, env)):
            return self.eval_node(node.then_val, env)
        return self.eval_node(node.else_val, env)

    def eval_ExprStmt(self, node: ExprStmt, env: Environment) -> Any:
        return self.eval_node(node.expr, env)

    def eval_Await(self, node: Await, env: Environment) -> Any:
        val = self.eval_node(node.value, env)
        if asyncio.iscoroutine(val):
            return asyncio.get_event_loop().run_until_complete(val)
        return val

    # ── Helpers ───────────────────────────────────────────────────────────────

    def _is_truthy(self, val: Any) -> bool:
        if val is None or val is False:
            return False
        if isinstance(val, (int, float)) and val == 0:
            return False
        if isinstance(val, (str, list, dict, set, tuple)) and len(val) == 0:
            return False
        return True

    def _to_str(self, val: Any) -> str:
        if val is None:
            return "shoonya"
        if val is True:
            return "satya"
        if val is False:
            return "asatya"
        return str(val)
```
