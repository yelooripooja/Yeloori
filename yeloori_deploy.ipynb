```python
import os
import shutil

# 1. Clean out any corrupted old build remnants
for folder in ['build', 'dist', 'yeloori.egg-info']:
    if os.path.exists(folder):
        shutil.rmtree(folder)

# 2. Force-create the exact layout structure safely
os.makedirs('yeloori/core', exist_ok=True)
os.makedirs('cli', exist_ok=True)

# 3. Create structural initialization scripts so Python treats them as packages
with open('yeloori/__init__.py', 'w', encoding='utf-8') as f:
    f.write('__version__ = "0.1.0"\\n')
with open('yeloori/core/__init__.py', 'w', encoding='utf-8') as f:
    f.write('')
with open('cli/__init__.py', 'w', encoding='utf-8') as f:
    f.write('')

# 4. Check for your uploaded yeloori.py code file and migrate it into the core engine path
if os.path.exists('yeloori.py'):
    with open('yeloori.py', 'r', encoding='utf-8') as src:
        core_code = src.read()
    with open('yeloori/core/engine.py', 'w', encoding='utf-8') as dest:
        dest.write(core_code)
    print("✓ Successfully migrated core language logic into yeloori/core/engine.py")
else:
    raise FileNotFoundError("CRITICAL ERROR: Please make sure your 'yeloori.py' file is uploaded to the Colab sidebar (📂) before running this cell!")

# 5. Re-generate the command-line interface runtime launcher script
cli_code = """import sys
import os
sys.path.insert(0, os.path.abspath(os.path.join(os.path.dirname(__file__), '..')))

from yeloori.core.engine import Lexer, Parser, Interpreter, LooriError

def run():
    if len(sys.argv) < 2:
        print("Yeloori Programming Language CLI v0.1.0")
        print("Usage: yeloori <filename.loori>")
        sys.exit(1)

    filename = sys.argv[1]
    if not os.path.exists(filename):
        print(f"Error: File '{filename}' not found.")
        sys.exit(1)

    with open(filename, 'r', encoding='utf-8') as f:
        source_code = f.read()

    try:
        lexer = Lexer(source_code, filename=filename)
        parser = Parser(lexer.tokenize(), filename=filename)
        interpreter = Interpreter()
        interpreter.execute(parser.parse())
    except LooriError as e:
        print(e)
    except Exception as e:
        print(f"Critical System Error: {e}")

if __name__ == '__main__':
    run()
"""

with open('cli/main.py', 'w', encoding='utf-8') as f:
    f.write(cli_code)

print("✓ All architecture modules generated perfectly. Compiling distribution binaries...")

# 6. Run the terminal compilation commands safely in the background
os.system("python3 -m build")

print("\\n" + "="*50 + "\\nBUILD COMPLETE! STARTING PYPI UPLOAD VETTING LIVE\\n" + "="*50)
```
